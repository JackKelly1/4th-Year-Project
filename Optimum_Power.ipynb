```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
N_channels = 40
data_path = os.getcwd()
```

```python
file_path = f'{data_path}/Data/optimum_power.csv'
df = pd.read_csv(file_path, header=None)
dataset = df.values
X = dataset[0,:]
Y = dataset[1,:]
plt.plot(X,Y)
plt.ylabel('Average GSNR')
plt.xlabel('Launch Power [dBm]')
plt.savefig(f'{data_path}/Figures/optimum_launch_power.svg', bbox_inches='tight', format='svg', dpi=1200)
plt.show()

ymax = max(Y)
X = X.tolist()
Y = Y.tolist()
xpos = Y.index(ymax)
xmax = X[xpos]
print(f'Optimum channel launch power is: {round(xmax, 2)}')
```
